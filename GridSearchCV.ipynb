```python
# above thres only one model model now what if we want to try many models
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = { 'svm' : {'model' : svm.SVC(gamma = 'auto') , 'params' : {'C' : [1,3,5,4,10] , 'kernel' : ['rbf','linear']}
                          },
                'random_forest' : {'model' : RandomForestClassifier() , 'params' : {'n_estimators' : [1,3,5,7,10]}
                                  },
                'logistic_regression' : {'model' : LogisticRegression(solver = 'liblinear' , multi_class = 'auto') , 'params' : {'C' : [1,3,5,7,10]}
                                        }
                }
```

```python
model_params
```

```python
for model_name , mp in model_params.items():
    print(model_name)
    print(mp) 
```

```python
for model_name , mp in model_params.items():
    print(mp)
```

```python
scores = []
for model_name , mp in model_params.items():
    print(mp)
    clf = GridSearchCV( mp['model'] , mp['params'] , cv = 5,return_train_score = False)
    clf.fit(digits.data,digits.target)
    scores.append({'model' : model_name
                  ,'best_score' : clf.best_score_
                  , 'best_params' : clf.best_params_})
scores
```

```python
scores
```

```python
pd.DataFrame(scores , columns = ['model','best_score','best_params'])
```
